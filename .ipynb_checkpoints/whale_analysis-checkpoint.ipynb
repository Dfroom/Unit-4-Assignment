### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
std_rolling_21_df = combined_returns_df.rolling(window=21).std()
# Plot the rolling standard deviation
std_rolling_21_df.plot(figsize=(20, 10))
```

### Calculate and plot the correlation

```python
# Calculate the correlation
correlation = combined_returns_df.corr()
# Display de correlation matrix
sns.heatmap(correlation)
```

```python
correlation
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate covariance of a single portfolio
algo_1_covariance = combined_returns_df["Algo 1"].rolling(window=60).cov(combined_returns_df["S&P TSX 60"])
# Calculate variance of S&P TSX
variance = combined_returns_df["S&P TSX 60"].rolling(window=60).var()
# Computing beta
algo_1_beta = algo_1_covariance / variance
# Plot beta trend
algo_1_beta.plot(figsize=(20, 10))
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios

ann_sharpe_ratio = (combined_returns_df.mean() * 252) / (std_combined * np.sqrt(252))
ann_sharpe_ratio.tail()
```

```python
# Visualize the sharpe ratios as a bar plot
ann_sharpe_ratio.plot.bar(figsize=(20, 10))
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Reading data from 1st stock
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

api = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2"
)
```

```python
# Referenced previous in class activities for this code
timeframe = "1Day"

# Set start and end datetimes of 1 year, between now and 365 days ago.
start_date = pd.Timestamp("2022-09-17", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2023-09-17", tz="America/New_York").isoformat()

# Set the stock tickers
tickers = ["LULU", "DIS", "AAPL", "SBUX"]
```

```python
# Reading data from 3rd stock
custom_portfolio_df = api.get_bars(
    tickers,
    timeframe,
    start=start_date,
    end=end_date,
).df
custom_portfolio_df.head()
```

```python
# Combine all stocks in a single DataFrame
LULU = custom_portfolio_df[custom_portfolio_df['symbol']=='LULU'].drop('symbol', axis=1)
DIS = custom_portfolio_df[custom_portfolio_df['symbol']=='DIS'].drop('symbol', axis=1)
AAPL = custom_portfolio_df[custom_portfolio_df['symbol']=='AAPL'].drop('symbol', axis=1)
SBUX = custom_portfolio_df[custom_portfolio_df['symbol']=='SBUX'].drop('symbol', axis=1)
custom_portfolio_df = pd.concat([LULU, DIS, AAPL, SBUX], axis=1, keys=["LULU", "DIS", "AAPL", "SBUX"])
custom_portfolio_df.head()
```

```python
# Reset Date index
df_custom_portfolio = pd.DataFrame()
for ticker in tickers:
    df_custom_portfolio[ticker] = custom_portfolio_df[ticker]["close"]
                                                              
df_custom_portfolio.index = df_custom_portfolio.index.date
df_custom_portfolio.head()
```

```python
# Reorganize portfolio data by having a column per symbol
```

```python
# Calculate daily returns
df_custom_portfolio = df_custom_portfolio.pct_change()
# Drop NAs
df_custom_portfolio.dropna(inplace=True)
# Display sample data
df_custom_portfolio.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/4, 1/4, 1/4, 1/4]

# Calculate portfolio return
weighted_returns = df_custom_portfolio.dot(weights)
# Display sample data
weighted_returns
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
df_custom_portfolio["Weighted Returns"] = weighted_returns
df_custom_portfolio
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
df_custom_portfolio.dropna(inplace=True)
df_custom_portfolio.isnull().sum()
```
